```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pygmt
```

```python
test_df = pd.read_csv("../data/test.2024.04.16.csv")
 # Five selected features
feature_col_names = ['F01', 'F07', 'F09', 'F10', 'F13']
# Get all testing examples that have values for the selected features
test_feats_df = test_df[~np.any(np.isnan(test_df[feature_col_names]), axis=1)]
```

```python
train_df = pd.read_csv("../data/train.2024.04.16.csv")
train_feats_df = train_df[~np.any(np.isnan(train_df[feature_col_names]), axis=1)]
```

```python
region = np.array([-114, -109, 37, 42])
fig = pygmt.Figure()
pygmt.config(FORMAT_GEO_MAP="ddd.x", MAP_FRAME_TYPE="plain", FONT='8p', FONT_ANNOT_PRIMARY='8p')
#fig.basemap(region=region, projection="M5/5")
fig.coast(region=region, projection="M6/6", land="white", borders=2, water="skyblue")
fig.plot(data=train_feats_df[train_feats_df["Tag"] == "eq"][['Elon', 'Elat']],style="c0.1c", pen="0.25p,black", fill='gray')
fig.plot(data=train_feats_df[train_feats_df["Tag"] == "mis"][['Elon', 'Elat']],style="c0.1c", pen="0.25p,black", fill='black')
fig.plot(data=train_feats_df[(train_feats_df["Tag"] == "mis") & (train_feats_df["Elat"] < 39.1)][['Elon', 'Elat']],style="c0.1c", pen="0.25p,black", fill='red')

fig.show()
```
